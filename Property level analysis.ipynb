### EDA

```python
import seaborn as sns
import matplotlib.pyplot as plt
import re
from pandas.plotting import table
```

```python
#improving readability through a line graph
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Bin occupancy rates
df['occupancy_rate_bin'] = df['occupancy_rate'].round()

# 2. Group and average revenue per bin
line_data = df.groupby('occupancy_rate_bin')['estimated_revenue_l365d'].mean().reset_index()

# 3. Fit linear regression trend line
X = line_data[['occupancy_rate_bin']]
y = line_data['estimated_revenue_l365d']
model = LinearRegression().fit(X, y)
trend = model.predict(X)

# 4. Plot line + trend line
plt.figure(figsize=(10, 6))
plt.plot(line_data['occupancy_rate_bin'], y, marker='o', label='Average Revenue', color='coral')
plt.plot(line_data['occupancy_rate_bin'], trend, linestyle='--', color='red', label='Trend Line')
plt.title('Average Revenue by Occupancy Rate with Trend Line')
plt.xlabel('Occupancy Rate (%)')
plt.ylabel('Average Revenue ()')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
corr = df[['occupancy_rate', 'estimated_revenue_l365d']].corr().iloc[0,1]
print(f"Correlation coefficient: {corr:.2f}")
```

Occupancy is a key driver of revenue, but not the only one.

- Other factors likely involved:
- Nightly price
- Seasonality
- Property type
- Location desirability

```python
def categorize_property(property_type):
    if 'Entire' in property_type:
        return 'Entire Home/Unit'
    elif 'Private room' in property_type:
        return 'Private Room'
    elif 'Shared room' in property_type:
        return 'Shared Room'
    elif any(word in property_type.lower() for word in ['hotel', 'hostel', 'resort', 'serviced apartment', 'bed and breakfast', 'aparthotel']):
        return 'Hotel-like'
    else:
        return 'Other'

# Apply to your DataFrame
df['property_category'] = df['property_type'].apply(categorize_property)
```

```python
# Group and sort
grouped = df.groupby('property_category').agg({
    'occupancy_rate': 'mean',
    'estimated_revenue_l365d': 'mean'
}).reset_index()

grouped = grouped.sort_values('occupancy_rate', ascending=False)

# Generate soft orange gradient with N unique colors
num_categories = len(grouped)
coral_palette = sns.light_palette("coral", n_colors=num_categories, reverse=True)  # Light to deeper orange

# Plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar plot with orange gradient
sns.barplot(
    x='property_category',
    y='estimated_revenue_l365d',
    data=grouped,
    palette=coral_palette,
    ax=ax1
)

# Left Y-axis: revenue
ax1.set_ylabel('Avg Revenue ()', color='black', fontsize=12)
ax1.set_xlabel('Property Category', color='black', fontsize=12)
ax1.tick_params(axis='y', labelcolor='black')
ax1.tick_params(axis='x', rotation=45, labelcolor='black')
ax1.set_xticklabels(ax1.get_xticklabels(), ha='right')

# Right Y-axis: occupancy rate line
ax2 = ax1.twinx()
sns.lineplot(
    x='property_category',
    y='occupancy_rate',
    data=grouped,
    ax=ax2,
    color='red',
    marker='o',
    linewidth=2
)

ax2.set_ylabel('Avg Occupancy Rate (%)', color='black', fontsize=12)
ax2.tick_params(axis='y', labelcolor='black')

# Title
plt.title('Avg Revenue and Occupancy Rate by Property Category', fontsize=14, color='black')
plt.tight_layout()
plt.show()
```

General Observations:
Positive trend overall between occupancy rate and estimated revenue.

"Entire Home/Unit" (orange) listings dominate both in number and revenue scale.

Other categories (e.g., Shared Room, Unique Stay, Hotel-like) remain clustered near the bottom revenue range, even at higher occupancy rates.

```python
#analysing review rating by yearly revenue
corr = df[['review_scores_rating', 'estimated_revenue_l365d']].corr().iloc[0, 1]
print(f"Correlation between reviews and annual revenue: {corr:.2f}")
```

Based off findings- we want to filter the entire homes/units that heave the highest revenue and occupancy rate. Once we pinpoint this we can begin sentiment analysis.

### Narrowing down our dataset

```python
# 1. Filter to only 'Entire Home/Unit'
entire_df = df[df['property_category'] == 'Entire Home/Unit'].copy()

# 2. Rank by revenue and occupancy rate (descending = higher rank)
entire_df['revenue_rank'] = entire_df['estimated_revenue_l365d'].rank(ascending=False, method='min')
entire_df['occupancy_rank'] = entire_df['occupancy_rate'].rank(ascending=False, method='min')

# 3. Keep only listings in top 20 for BOTH
top_10_both = entire_df[
    (entire_df['revenue_rank'] <= 20) &
    (entire_df['occupancy_rank'] <= 20)
].sort_values(by='estimated_revenue_l365d', ascending=False)
```

```python
display_cols = ['name', 'estimated_revenue_l365d', 'occupancy_rate', 'id', 'accommodates', 'neighbourhood_cleansed']
top_10_both[display_cols]
```

These are the 10 listings that fulfil both conditions- highest in occupancy rates, and highest in yearly revenue.

```python
top = top_10_both.sort_values('estimated_revenue_l365d', ascending=True)

plt.figure(figsize=(10, 6))
plt.barh(top['name'], top['estimated_revenue_l365d'], color='coral')
plt.xlabel('Annual Revenue ()')
plt.title('Top 10 Listings by Revenue & Occupancy Rate (Entire Home/Unit)')
plt.tight_layout()
plt.show()
```
